```python
import matplotlib.pyplot as plt
```

```python
from PIL import Image, ImageDraw, ImageFont

# Open the image
img = Image.open("C:\\Datasets\\sivaji_dataanalysis.jpg")

# Create a Draw Object
DRAW = ImageDraw.Draw(img)

# Preparing the Text Element
TEXT = """*GEN_AI_INTERNSHIP"""

TEXT2 = '''** ARE 
U REGISTER 
TO KANAV SIR
GEN AI INTERNSHIP'''

TEXT3 = """ME ALREADY 
LISTENING 
TO THAT CLASS"""

# Customize font type and font style
CUST_FONT = ImageFont.truetype(r'C:\Users\System-Pc\Desktop\arial.ttf', 40)

# Add text to the image
DRAW.text((100, 10), text=TEXT, fill=(20, 250, 0), font = CUST_FONT)

DRAW.text((120, 310), text=TEXT2, fill=(250, 250, 0), font = CUST_FONT)

DRAW.text((200, 550), text=TEXT3, fill=(250, 250, 0), font = CUST_FONT)

# Display the image
plt.imshow(img)
img.save(r"C:\\Datasets\\sivaji.jpg")
```
